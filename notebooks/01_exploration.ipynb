```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Imports successful!")
```

```python
# Load first 10,000 rows to safely inspect structure
df_sample = pd.read_csv(
    '../data/raw/MUP_DPR_RY25_P04_V10_DY23_NPIBN.csv',
    nrows=10000,
    low_memory=False
)

print(f"Sample shape: {df_sample.shape}")
print(f"\nColumn names:")
for col in df_sample.columns:
    print(f"  {col}")
```

```python
# Check data types
print("=== DATA TYPES ===")
print(df_sample.dtypes)

print("\n=== NULL VALUE COUNTS ===")
null_counts = df_sample.isnull().sum()
null_pct = (null_counts / len(df_sample) * 100).round(2)
null_summary = pd.DataFrame({
    'null_count': null_counts,
    'null_pct': null_pct
})
print(null_summary[null_summary['null_count'] > 0])
```

```python
# How many unique providers in the sample?
print(f"Unique NPIs: {df_sample['Prscrbr_NPI'].nunique()}")
print(f"Unique drugs (generic): {df_sample['Gnrc_Name'].nunique()}")
print(f"Unique drugs (brand): {df_sample['Brnd_Name'].nunique()}")

print(f"\nTop 10 states by row count:")
print(df_sample['Prscrbr_State_Abrvtn'].value_counts().head(10))

print(f"\nTop 15 provider specialties:")
print(df_sample['Prscrbr_Type'].value_counts().head(15))
```

```python
# Load full dataset filtered to Texas only
chunks = []
chunk_size = 50000

for chunk in pd.read_csv(
    '../data/raw/MUP_DPR_RY25_P04_V10_DY23_NPIBN.csv',
    chunksize=chunk_size,
    low_memory=False
):
    texas_chunk = chunk[chunk['Prscrbr_State_Abrvtn'] == 'TX']
    chunks.append(texas_chunk)

df_texas = pd.concat(chunks, ignore_index=True)

print(f"Texas dataset shape: {df_texas.shape}")
print(f"Unique Texas providers: {df_texas['Prscrbr_NPI'].nunique()}")
print(f"Unique drugs: {df_texas['Gnrc_Name'].nunique()}")
```

```python
df_texas.to_csv('../data/processed/texas_part_d_2023.csv', index=False)
print("Texas data saved!")
print(f"File saved to data/processed/texas_part_d_2023.csv")
```
